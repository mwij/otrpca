```python
def parallel_apply(func1d, arr, n, positive):

    chunks = [(func1d, rank, n, positive) for rank in arr]

    pool = multiprocessing.Pool()
    individual_results = pool.map(unpacking_apply, chunks)
    # Freeing the workers:
    pool.close()
    pool.join()

    return(individual_results)
```

```python
def unpacking_apply(chunks):

    
    func1d, rank, n, positive = chunks
    
    return(func1d(rank[0], n, rank[1], positive ))
```

```python
rhos = [.51, .4, .3, .2, .1, .01]

n = 6

for n in [10, 20, 50]: 
    ranks = [i + 1 for i in range(int(n/2))]
    results = []

    for j in range(len(rhos)):
        results = results + [ (ranks[i], rhos[j]) for i in range(len(ranks)) ]
        
    results = parallel_apply(vary_rank, results, n, False)

    plt.figure(dpi=300)

    for i in range(len(rhos)):
        plt.plot(results[i*n: n*(i + 1)], label='rho = ' + str(rhos[i]))

    plt.xlabel('Tensor Tubal Rank')
    plt.ylabel('Low Rank Recovery Error')
    plt.legend(title = 'n = ' + str(n))
    plt.ylim(0, 1)

    plt.savefig('Pictures/L'+str(n))
    plt.show()
```
